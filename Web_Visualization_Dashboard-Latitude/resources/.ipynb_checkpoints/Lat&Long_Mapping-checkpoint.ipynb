```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csvpath = "citySample_dataframe.csv"
cities = pd.read_csv(csvpath)
cities.head()
```

```python
sns.set_style("darkgrid")
ax = cities.plot(x="Longitude", y="Latitude", kind="scatter", figsize=(11,8), alpha=0.5)
ax.set_ylim(-90,90)
# draw line at y=0
plt.axhline(0, color='crimson', alpha=0.25)
plt.title("City Sample Location Map", fontsize=24)
# labels and font size
plt.xlabel("Longitude", fontsize=18)
plt.ylabel("Latitude", fontsize=18)
    # tick sizes
ax.tick_params(labelsize=15)
#Save figure as "CitySampleLocationMap.png"
plt.savefig("CitySampleLocationMap.png")
plt.show()

```
